```python
merged_ratings = ratings.merge(books, on = 'ISBN')
```

```python
merged_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_change = merged_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_change.rename(columns={'Book-Rating':'ratings_number'},inplace=True)
num_rating_change
```

```python
avg_rating = merged_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating
```

```python
popular_books = num_rating_change.merge(avg_rating,on='Book-Title')
popular_books
```

```python
popular_books = popular_books[popular_books['ratings_number']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_books = popular_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','ratings_number','avg_rating']]
```

```python
x = merged_ratings.groupby('User-ID').count()['Book-Rating'] > 200
like_users = x[x].index
```

```python
filtered_rating = merged_ratings[merged_ratings['User-ID'].isin(like_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
most_famous_books = y[y].index
```

```python
end_ratings = filtered_rating[filtered_rating['Book-Title'].isin(most_famous_books)]
```

```python
piv_tab = end_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```
