```python
import pandas as pd
import numpy as np
import scipy as sp
from sklearn.metrics.pairwise import cosine_similarity
import operator
```

```python
rating = pd.read_csv('/content/drive/MyDrive/archivee/rating.csv')

# DataFrame'i incele
rating.head()


anime = pd.read_csv('/content/drive/MyDrive/archivee/anime.csv')

# DataFrame'i incele
anime.head()
```

```python
rating.rating.replace(-1, np.nan, inplace=True)
rating.head()

anime_special = anime[anime['type']=='Special']
anime_special.head()
```

```python
# Verileri anime_id sütununa göre birleştirme
merged_data = rating.merge(anime_special, on='anime_id', suffixes= ['_user', ''])
merged_data.head()
```

```python
merged_data = merged_data[['user_id', 'name', 'rating_user']]
merged_new = merged_data[merged_data.user_id <= 10000]
merged_new.head()
```

```python
pivot_tab = merged_new.pivot_table(index=['name'], columns=['user_id'], values='rating_user')
print(pivot_tab.shape)
pivot_tab.head(20)
```

```python
#Veriyi normalize et
pivot_norm = pivot_tab.apply(lambda x: (x-np.mean(x))/(np.max(x)-np.min(x)), axis=1)

```

```python
pivot_norm.fillna(0, inplace=True)
pivot_norm = pivot_norm.loc[:, (pivot_norm != 0).any(axis=0)]

pivot_norm.head()
```

```python
pivot_spr = sp.sparse.csr_matrix(pivot_norm.values)
```

```python
sim_item = cosine_similarity(pivot_spr.T)
sim_user = cosine_similarity(pivot_spr)
```

```python
# sim_item matrisini 1192x1192 boyutuna sınırla
sim_item_df = pd.DataFrame(sim_item[:1192, :1192], index=pivot_norm.index, columns=pivot_norm.index)

# sim_user matrisini 1192x1192 boyutuna sınırla
sim_user_df = pd.DataFrame(sim_user[:1192, :1192], index=pivot_norm.index[:1192], columns=pivot_norm.columns[:1192])
```

```python
def best_animes_names(anime):
    num = 1
    print(f'Similar animes to {anime}:\n')
    for name in sim_item_df.sort_values(by = anime, ascending = False).index[1:11]:
        print(f'Number. {num}: {name}')
        num +=1
```

```python
best_animes_names('Working!!!: Lord of the Takanashi')
```

```python
def most_sim_users(user):

    if user not in sim_user_df.columns:
        return(f'No knowledge available on user {user}')

    print('High Similar Users:\n')
    sim_val = sim_user_df.sort_values(by=user, ascending=False).loc[:,user].tolist()[1:11]
    sim_usr = sim_user_df.sort_values(by=user, ascending=False).index[1:11]
    zip_list = zip(sim_usr, sim_val)
    for user_name, similarity in zip_list:
        print(f'User Name #{user_name}, Similarity: {similarity:.2f}')
```

```python
most_sim_users(3)
```

```python
def similar_user_recs(user):

    if user not in pivot_norm.T.columns:
        return('No data available on user {}'.format(user))

    sim_users = sim_user_df.sort_values(by=user, ascending=False).index[1:11]
    best = []
    most_common = {}

    for i in sim_users:
        max_score = pivot_norm.loc[:, i].max()
        best.append(pivot_norm[pivot_norm.loc[:, i]==max_score].index.tolist())
    for i in range(len(best)):
        for j in best[i]:
            if j in most_common:
                most_common[j] += 1
            else:
                most_common[j] = 1
    sorted_list = sorted(most_common.items(), key=operator.itemgetter(1), reverse=True)
    return sorted_list[:5]

```

```python
similar_user_recs(3)
```

```python
def predicted_rating(anime_name, user):
    sim_users = sim_user_df.sort_values(by=user, ascending=False).index[1:1000]
    user_values = sim_user_df.sort_values(by=user, ascending=False).loc[:,user].tolist()[1:1000]
    rating_list = []
    weight_list = []
    for j, i in enumerate(sim_users):
        rating = pivot_tab.loc[i, anime_name]
        similarity = user_values[j]
        if np.isnan(rating):
            continue
        elif not np.isnan(rating):
            rating_list.append(rating*similarity)
            weight_list.append(similarity)
    return sum(rating_list)/sum(weight_list)
```
